```python
from PIL import Image as Im
from PIL import ImageFont
from PIL import ImageDraw
from PIL import ImageTk
import random
import os
import string
from tqdm import tqdm
import net as nn
import torch
import cv2
import numpy as np
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
```

```python
net = nn.Net()
net.load_weights('trained_data.npy')

files = []

keep_going = True
```

```python
def selectDir():
    global files, root
    files = []
    directory = filedialog.askdirectory(initialdir = '', title = 'Select Directory')
    if directory:
        for (root_dir, dirs, dir_files) in os.walk(directory):
            for file in dir_files:
                if file[-4:] == '.jpg':
                    files.append(os.path.join(root_dir, file))
    if len(files) > 0:
        startButton['state'] = NORMAL
            
    
def selectImage():
    global files
    files = [filedialog.askopenfilename(initialdir = '', title = 'Select Image')]
    
    if files[0]:
        startButton['state'] = NORMAL
    
def start():
    global files, keep_going
    
    startButton['state'] = DISABLED
    
    maxLength = len(files)
    progressBar = ttk.Progressbar(progressBarFrame, orient=HORIZONTAL,length=500,mode='determinate')
    progressBar.pack()
    
    for file in files:
        progressBar['value'] = files.index(file) / maxLength * 100
        
        image = Im.open(file)
        
        root.one = one = ImageTk.PhotoImage(image)
        canvas.create_image((image.width / 2, image.height / 2), image = one, anchor = CENTER)
        root.update()
        
        rect = None
        
        for x in range(0, image.width, 32):
            for y in range(0, image.height, 16):
                
                input_image = image.crop((x, y, x + 64, y + 32)).save('tempImage.jpg')
                input_image = cv2.imread('tempImage.jpg', cv2.IMREAD_GRAYSCALE)
                input_image = cv2.resize(input_image, (128, 64))
                input_image = torch.Tensor(input_image).view(-1, 128, 64)/255.0
                input_image = input_image.to(nn.device).view(-1, 1, 128, 64)
                
                net_out = torch.argmax(net(input_image))
                
                if net_out == 0:
                    canvas.create_rectangle(x, y, x + 64, y + 32, outline = 'green')
                    
                    root.update()
                    
                if not keep_going:
                    return None
    
    progressBar.pack_forget()
    startButton['state'] = NORMAL
    
def stop():
    global keep_going
    keep_going = False
                
```

```python
root = Tk()
root.geometry('1024x1024')

title = Label(root, text = 'Dicom Text Remover', font = ('Fonts/Arial.ttf', 32))
title.pack(padx = 50, pady = 50)

selectFrame = Frame(root)
selectFrame.pack()

selectDirButton = Button(selectFrame, text = 'Select Directory', command = selectDir, bg = 'gray', activebackground = 'light gray')
selectDirButton.pack(padx = 25, pady = 25, side = LEFT)

selectImageButton = Button(selectFrame, text = 'Select Image', command =  selectImage, bg = 'gray', activebackground = 'light gray')
selectImageButton.pack(padx = 25, pady = 25, side = RIGHT)

start_end_button_frame = Frame(root)
start_end_button_frame.pack(pady = 25, side = TOP)
startButton = Button(start_end_button_frame, text = 'Start', command = start, state = DISABLED, bg = 'gray', activebackground = 'light gray')
startButton.pack(padx = 25, side=LEFT)
endButton = Button(start_end_button_frame, text = 'Stop', command = stop, state = DISABLED, bg = 'gray', activebackground = 'light gray')
endButton.pack(padx = 25, side = LEFT)

canvasFrame = Frame(root, borderwidth = 2, width = 514, height = 514, relief = SOLID)
canvasFrame.pack(pady=25, side = TOP)
canvas = Canvas(canvasFrame)
canvas.config(width = 512, height = 512)
canvas.pack()

progressBarFrame = Frame(root)
progressBarFrame.pack(pady = 25, side = TOP)

error_msg = Label(root)
error_msg.pack(pady = 25, side = TOP)


root.mainloop()
```
